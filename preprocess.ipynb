```python
import pandas as pd
import numpy as np
```

```python
#Preprocessing Rice dataset
data=pd.read_csv("rice.csv")
data["label"]=data["label"].replace("Cammeo", 0) 
data["label"]=data["label"].replace("Osmancik", 1)
data.to_csv("preprocessed_datasets/rice_processed.csv", index=False) 
```

```python
#Preprocessing Parkinsons dataset
#Needs no change as it is completely a numerical dataset
data=pd.read_csv("parkinsons.csv")
data.to_csv("preprocessed_datasets/parkinsons_processed.csv", index=False)
```

```python
#Preprocessing Digits dataset

from sklearn import datasets

digits=datasets.load_digits(return_X_y=True)
data=pd.DataFrame(digits[0])

#Adding the class labels (0-9) as the last column
data["label"]=digits[1]
data.to_csv("preprocessed_datasets/digits_processed.csv", index=False)
```

```python
#Preprocessing Credit Approval dataset

data=pd.read_csv("credit_approval.csv")

#One-hot encoding all categorical columns
data=pd.get_dummies(data, columns=["attr1_cat", "attr4_cat", "attr5_cat", "attr6_cat", 
                                      "attr7_cat", "attr9_cat", "attr10_cat", "attr12_cat", 
                                      "attr13_cat"]) 

#Moving label to the last column
label=data.pop("label")
data["label"] = label

data = data.astype(float)
data.to_csv("preprocessed_datasets/credit_processed.csv", index=False)
```

```python
##Preprocessing Student Dropout dataset
data=pd.read_csv("student_dropout.csv", sep=";")

# Converting target labels to numbers
data["Target"]=data["Target"].replace("Dropout", 0)
data["Target"]=data["Target"].replace("Enrolled", 1)
data["Target"]=data["Target"].replace("Graduate", 2)

#One-hot encoding categorical columns
data=pd.get_dummies(data, columns=["Marital status", "Application mode", 
                                      "Application order", "Course", "Nacionality",
                                      "Previous qualification", "Mother's qualification", 
                                      "Father's qualification", "Mother's occupation", 
                                      "Father's occupation"])

#Moving target to the last column
target = data.pop("Target")
data["Target"] = target

data = data.astype(float)  
data.to_csv("preprocessed_datasets/student_dropout_processed.csv", index=False)
```
